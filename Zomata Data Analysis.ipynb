# IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

# CONVERT THE DATA TYPE OF COLUMN - RATE

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# TYPE OF RESTAURANT

```python
dataframe
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

```python
sns.countplot(
    x='listed_in(type)',
    data=dataframe,
    hue='listed_in(type)',
    palette='Set1',
    legend=False
)

plt.xlabel("Type of Restaurant")
plt.xticks(rotation=45)
plt.show()
```

CONCLUSION : majority of the restaurant falls in the dining category.

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

plt.plot(grouped_data.index, grouped_data.values,
         color="green", marker="o")

plt.xlabel("Type of Restaurant", color="blue", size=15)
plt.ylabel("Total Votes", color="blue", size=15)
plt.xticks(rotation=45)
plt.show()
```

CONCLUSION : Dining restaurant has received maximum votes.

```python
dataframe
```

```python
plt.hist(dataframe['rate'], bins = 5)
plt.title('Rating Distribution')
plt.show()
```

CONCLUSION: The majority ratings are between 3.5 to 4.

# WHICH MODE RECEIVES MAXIMUM RATING ?

```python
dataframe
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

CONCLUSION : Offline receives low rating in comparion to online order.

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

CONCLUSION: Dining restaurant receive more offline orders, cafe receives more online orders. 
This shows that customers prefer more offline oredrs at dinning restaurants and more online orders at cafes.
